```python
import os
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import sys 
sys.path.append("/media/moon/Data/study/birds/exps/convergence2/")

file = 'data_cnn_fashionmnist_lr'
with open('./data/{}'.format(file)) as f:
    data = f.read()

def process_data(data):
    lines = data.split('\n')
    result = []
    
    tails = []
    t = 0
    for line in lines:    
        if '.' not in line:
            continue
        ele_list = line.split()
        if t % 4 == 0:
            tails.append(ele_list[7].split('_')[-1][4:])
        t = t+1
        result.append(ele_list[6])
    if len(tails) == 1 and tails[0] == '0':
        tails = ['$\\infty$', '$\\infty$']
    
    result = [float(i) for i in result]
    results = [result[8*i:8*i+8] for i in range(len(result)//4)]
    results = [item for item in results if item != []]
    print(results)
    return results

def plotcurve(data):    
    bar_width = 0.4
    titles = ['$C=1$, $K=5$', '$C=1$, $K=20$', '$C=1$, $K=50$', '$C=2$, $K=5$', '$C=2$, $K=20$', '$C=2$, $K=50$']
    fig, axs = plt.subplots(2, 3, figsize=(6, 4))
    
    x = np.arange(5)
    for i in range(2):
        for j in range(3):
            y1, y2 = data[i*3+j][:4], data[i*3+j][4:]
            y1_line, y2_line = [None] + y1, y2 + [None]
            y1, y2 = [0] + y1, y2 + [0]
            y1_mark, y2_mark = np.array(y1) == max(y1), np.array(y2) == max(y2)

            axs[i,j].bar(x + bar_width / 2, y1, bar_width, label='PFL', color='lightblue', alpha=0.7)
            axs[i,j].bar(x - bar_width / 2, y2, bar_width, label='SFL', color='lightcoral', alpha=0.7)
            axs[i,j].plot(x + bar_width / 2, y1_line, linestyle=':', color='C0', linewidth=2, marker='*', markevery=y1_mark, markersize=9)
            axs[i,j].plot(x - bar_width / 2, y2_line, linestyle=':', color='C3', linewidth=2, marker='*', markevery=y2_mark, markersize=9)
            print(i,j, titles[i*3+j])
            axs[i,j].set_xticks(x)
            axs[i,j].set_xticklabels([-2.5, -2.0, -1.5, -1.0, -0.5])
            axs[i,j].set_title(titles[i*3+j], fontsize=12)
            if i == 1:
                axs[i,j].set_xlabel('Learning rate ($\log_{10}$)')
            if j == 0:
                axs[i,j].set_ylabel('Test Top1 Accuracy (%)')
            if i == 0 and j == 2:
                axs[i,j].legend(loc=1, ncols=1, fontsize=10)
            #fig.suptitle('Figure Title', fontsize=16)
    plt.tight_layout()
    filename = 'lr_{}'.format(file[5:-3])
    print(filename)
    #plt.savefig('../figs/{}.pdf'.format(filename), bbox_inches='tight', dpi=300)


def main():
    results = process_data(data)
    plotcurve(results)

if __name__ == '__main__':
    main()
```
