```python
import matplotlib.pyplot as plt
import numpy as np

# Let's assume we have the following data for the bars (positive and negative values)
data = [2, -3, 5, -1, 4, -6, 3]

# We need positive and negative values separated to plot them symmetrically around zero
positive_values = np.abs(np.array(data))
negative_values = -np.array(data)

# The positions of the bars
bar_positions = np.arange(len(data))

# Create figure and axes
fig, ax = plt.subplots()

# Plot positive bars
ax.bar(bar_positions, positive_values, bottom=negative_values, color='b')

# Since we've already plotted the positive bars over the negative ones,
# there's no need to plot the negative bars separately.
# However, if you wish to also show the negative bars explicitly:
# ax.bar(bar_positions, negative_values, color='r')

# Set limits so that it's centered at zero
ax.set_xlim(-1, len(data))
ax.set_ylim(min(min(positive_values), min(negative_values)), max(max(positive_values), max(negative_values)))

# Add x-axis tick labels (optional)
#plt.xticks(bar_positions, ['Category1', 'Category2', 'Category3', 'Category4', 'Category5', 'Category6', 'Category7'])

# Show gridlines
ax.grid(True)

# Show y-axis tick marks on both sides
ax.yaxis.tick_left()
ax.yaxis.tick_right()

# Add labels
ax.set_xlabel('Categories')
ax.set_ylabel('Values')

# Display the plot
plt.show()
```
